###コードレビューの流れ

0. 大元のRepository(以下 upstream)をforkする(以下, forkしたものをRemote).
0. Remoteのレポジトリをクローン.
0. branchの作成
0. 課題をやる
    00. 課題を読む
    00. 課題に沿ったプログラミングを行う(commitの粒度は適当に squashを使ってまとめても良い)
        00. コード作成
        00. commit
        00. 動作確認
        00. commit
        00. リファクタリング
        00. commit
        00. Remoteへpush
    00. gitlabでupstream に pull request(merge request)を送る.
    00. コードのレビューが帰ってきたらB.に戻る.
